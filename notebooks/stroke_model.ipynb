```python
import tf2onnx
import tensorflow as tf

# Pehle SavedModel format mein save karo
model = tf.keras.models.load_model('model.h5')
model.export('saved_model_dir') # Use model.export() for SavedModel format

# Ab SavedModel se ONNX banao
!python -m tf2onnx.convert --saved-model saved_model_dir --output model.onnx --opset 13

print("Done!")
```
